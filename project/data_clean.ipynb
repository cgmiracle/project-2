```python
import pandas as pd
import os
import numpy as np
```

```python
playerstats = pd.read_csv("playstats.csv")
playerstats.head()
```

```python
#Cleaning Data
#Selecting only wanted columns
df = playerstats.filter(["Player","Pos", "MP","3P", "2P", "2PA",'FT','FTA','ORB','DRB','AST','STL','BLK','TOV','PF','PTS'])
df.head()
```

```python
df.dtypes
```

```python
#Multplying columns by fantasy point values
df["3P"] = .5 * df["3P"]
df["2P"] = .5 * df["2P"]
df["2PA"] = (-0.225) * df["2PA"]
df["FT"] = .5 * df["FT"]
df["FTA"] = (-0.4) * df["FTA"]
df["ORB"] = 2.5 * df["ORB"]
df["DRB"] = 2 * df["DRB"]
df["AST"] = 2 * df["AST"]
df["STL"] = 3 * df["STL"]
df["BLK"] = 3 * df["BLK"]
df["TOV"] = (-2.5) * df["TOV"]
df["PF"] = (-1.5) * df["PF"]
df["PTS"] = 1 * df["PTS"]
```

```python
column_list = list(df)
print(column_list)
```

```python
column_list.remove("Player")
column_list.remove("Pos")
column_list.remove("MP")
print(column_list)
```

```python
df["Fantasy Points"] = df[column_list].sum(axis=1)
df.head()
```
